```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from google.colab import files

# Allow the user to upload the .csv file
uploaded = files.upload()

# Get the uploaded file name (assuming the user uploads the correct CSV)
filename = next(iter(uploaded))  # Get the name of the uploaded file

# Load dataset
data = pd.read_csv(filename)

# Check for missing values and fill them with the mean (for numeric columns)
data.fillna(data.mean(numeric_only=True), inplace=True)

# Define independent (X) and dependent (y) variables
X = data[['year']]
y = data['per capita income (US$)']

# Train Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predict per capita income for 2020
year_2020 = np.array([[2020]])
income_2020 = model.predict(year_2020)

print(f"Predicted per capita income in 2020: ${income_2020[0]:.2f}")

# Plot regression line
plt.scatter(X, y, color='blue', label="Actual Data")
plt.plot(X, model.predict(X), color='red', linewidth=2, label="Regression Line")
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Canada Per Capita Income Prediction')
plt.legend()
plt.show()
```

```python
from google.colab import files

# Upload CSV file
uploaded = files.upload()
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv('canada_per_capita_income.csv')

# Check column names
print("Column Names:", df.columns)

# Rename columns if needed
df.rename(columns={'per capita income (US$)': 'Income'}, inplace=True)

# Remove any NaN values
df.dropna(inplace=True)

plt.xlabel('Year')
plt.ylabel('Income')

# Use correct column names
plt.scatter(df['year'], df['Income'], color='red', marker='+')

# Prepare input feature
X = df[['year']]  # Independent variable
y = df['Income']  # Dependent variable

# Train linear regression model
reg = linear_model.LinearRegression()
reg.fit(X, y)

# Predict future incomes
year_2020 = reg.predict([[2020]])[0]
year_2027 = reg.predict([[2027]])[0]

print(f'Year=2020, Predicted Income=${year_2020:,.2f}')
print(f'Year=2027, Predicted Income=${year_2027:,.2f}')
print("-------------------------------------------------------------------------------------------")
print(f'Coefficient (m): {reg.coef_[0]:,.2f}')
print("-------------------------------------------------------------------------------------------")
print(f'Intercept (b): {reg.intercept_:,.2f}')
print("-------------------------------------------------------------------------------------------")

# Plot regression line
plt.plot(df['year'], reg.predict(X), color='blue')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('homeprices_Multiple_LR.csv')
print(df.head())

# Check for missing values and fill with column means
print(df.isnull().sum())
df = df.fillna({
    'area': df['area'].mean(),
    'bedrooms': df['bedrooms'].mean(),
    'age': df['age'].mean(),
    'price': df['price'].mean()
})
print(df.isnull().sum())
print("---------------------------------")

# Scatter plot (using 'area' vs 'price' as an example)
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area, df.price, color='red', marker='+')

# Prepare input features (multiple independent variables)
X = df[['area', 'bedrooms', 'age']]  # Independent variables
y = df['price']  # Dependent variable

# Train multiple linear regression model
reg = linear_model.LinearRegression()
reg.fit(X, y)

# Predict future prices (example: area=3000, bedrooms=4, age=10)
# Use a DataFrame with column names to avoid sklearn warning
input_data = pd.DataFrame([[3000, 4, 10]], columns=['area', 'bedrooms', 'age'])
prediction = reg.predict(input_data)[0]

print(f'Area=3000, Bedrooms=4, Age=10, Predicted Price=${prediction:,.2f}')
print("-------------------------------------------------------------------------------------------")
print(f'Coefficients (m): {reg.coef_}')
print("-------------------------------------------------------------------------------------------")
print(f'Intercept (b): {reg.intercept_:,.2f}')
print("-------------------------------------------------------------------------------------------")

# Plot regression line (using 'area' as the x-axis for visualization, holding others constant)
area_range = np.linspace(df.area.min(), df.area.max(), 100)
fixed_bedrooms = df['bedrooms'].mean()
fixed_age = df['age'].mean()
X_pred = pd.DataFrame({
    'area': area_range,
    'bedrooms': [fixed_bedrooms] * 100,
    'age': [fixed_age] * 100
})
price_pred = reg.predict(X_pred)

plt.plot(area_range, price_pred, color='blue')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('1000_Companies.csv')
print("Dataset Head:")
print(df.head())
print("\nDataset Shape:", df.shape)

# Check for missing values and fill with column means
print("\nMissing Values Before:")
print(df.isnull().sum())
df = df.fillna({
    'R&D Spend': df['R&D Spend'].mean(),
    'Administration': df['Administration'].mean(),
    'Marketing Spend': df['Marketing Spend'].mean(),
    'Profit': df['Profit'].mean()
})
print("\nMissing Values After:")
print(df.isnull().sum())
print("---------------------------------")

# Scatter plot (using 'R&D Spend' vs 'Profit')
plt.xlabel('R&D Spend')
plt.ylabel('Profit')
plt.scatter(df['R&D Spend'], df['Profit'], color='red', marker='+')

# Prepare input features
X = df[['R&D Spend', 'Administration', 'Marketing Spend']]
y = df['Profit']

# Train multiple linear regression model
reg = linear_model.LinearRegression()
reg.fit(X, y)

# Predict profit
input_data = pd.DataFrame([[100000, 120000, 300000]],
                          columns=['R&D Spend', 'Administration', 'Marketing Spend'])
prediction = reg.predict(input_data)[0]

print(f'R&D Spend=100000, Administration=120000, Marketing Spend=300000, Predicted Profit=${prediction:,.2f}')
print("-------------------------------------------------------------------------------------------")
print(f'Coefficients (m): {reg.coef_}')
print("-------------------------------------------------------------------------------------------")
print(f'Intercept (b): {reg.intercept_:,.2f}')
print("-------------------------------------------------------------------------------------------")
print(f'R Score: {reg.score(X, y):.4f}')  # Goodness of fit

# Plot regression line (R&D Spend vs Profit, others fixed)
rd_range = np.linspace(df['R&D Spend'].min(), df['R&D Spend'].max(), 100)
fixed_admin = df['Administration'].mean()
fixed_marketing = df['Marketing Spend'].mean()
X_pred = pd.DataFrame({
    'R&D Spend': rd_range,
    'Administration': [fixed_admin] * 100,
    'Marketing Spend': [fixed_marketing] * 100
})
profit_pred = reg.predict(X_pred)

plt.plot(rd_range, profit_pred, color='blue')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

# Function to convert textual experience to numeric
def convert_experience(exp):
    if pd.isna(exp):  # Handle NaN
        return np.nan
    exp = str(exp).lower()
    num_dict = {
        'zero': 0, 'one': 1, 'two': 2, 'three': 3, 'four': 4,
        'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9, 'ten': 10
    }
    return num_dict.get(exp, np.nan)  # Return NaN if not found

# Load the dataset
df = pd.read_csv('hiring.csv')
print("Dataset Head:")
print(df.head())
print("\nDataset Shape:", df.shape)

# Convert 'experience' column to numeric
df['experience'] = df['experience'].apply(convert_experience)

# Check for missing values and fill with column means
print("\nMissing Values Before:")
print(df.isnull().sum())

# Fill NaN in experience with the mean of numeric values
df['experience'] = df['experience'].fillna(df['experience'].mean())
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].mean())
df['interview_score(out of 10)'] = df['interview_score(out of 10)'].fillna(df['interview_score(out of 10)'].mean())
df['salary($)'] = df['salary($)'].fillna(df['salary($)'].mean())

print("\nMissing Values After:")
print(df.isnull().sum())
print("---------------------------------")

# Scatter plot (using 'experience' vs 'salary($)')
plt.xlabel('Experience')
plt.ylabel('Salary ($)')
plt.scatter(df['experience'], df['salary($)'], color='red', marker='+')

# Prepare input features
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

# Train multiple linear regression model
reg = linear_model.LinearRegression()
reg.fit(X, y)

# Predict salary (example: experience=5, test_score=8, interview_score=7)
input_data = pd.DataFrame([[5, 8, 7]],
                          columns=['experience', 'test_score(out of 10)', 'interview_score(out of 10)'])
prediction = reg.predict(input_data)[0]

print(f'Experience=5, Test Score=8, Interview Score=7, Predicted Salary=${prediction:,.2f}')
print("-------------------------------------------------------------------------------------------")
print(f'Coefficients (m): {reg.coef_}')
print("-------------------------------------------------------------------------------------------")
print(f'Intercept (b): {reg.intercept_:,.2f}')
print("-------------------------------------------------------------------------------------------")
print(f'R Score: {reg.score(X, y):.4f}')  # Goodness of fit

# Plot regression line (using 'experience' as x-axis, others fixed)
exp_range = np.linspace(df['experience'].min(), df['experience'].max(), 100)
fixed_test_score = df['test_score(out of 10)'].mean()
fixed_interview_score = df['interview_score(out of 10)'].mean()
X_pred = pd.DataFrame({
    'experience': exp_range,
    'test_score(out of 10)': [fixed_test_score] * 100,
    'interview_score(out of 10)': [fixed_interview_score] * 100
})
salary_pred = reg.predict(X_pred)

plt.plot(exp_range, salary_pred, color='blue')
plt.show()
```
